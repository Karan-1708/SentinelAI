```python
# Feature correlation heatmap (top 20 features)
if 'df' in dir():
    numeric = df.select_dtypes(include=[np.number])
    corr = numeric.corr().abs()
    top_features = corr.nlargest(20, corr.columns[0]).index
    
    fig, ax = plt.subplots(figsize=(14, 12))
    sns.heatmap(corr.loc[top_features, top_features], ax=ax, cmap='RdBu_r', center=0, fmt='.2f', annot=False)
    ax.set_title('Feature Correlation Matrix (Top 20)')
    plt.tight_layout()
    plt.show()
```
